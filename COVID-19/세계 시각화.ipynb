```python
import json
import pandas as pd
import folium
print(folium.__version__)
```

```python
df = pd.read_csv('TC_M.csv')
```

```python
df
```

```python
geo_path = 'TM.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python

map = folium.Map(location=[36, 127], zoom_start=1)
map.choropleth(geo_data=geo_str, data=df,
              columns=['iso_code','total_cases_per_million'],
              key_on='properties.iso3',
              fill_color='YlGn',
              legend_name='iso_code')
map
```
